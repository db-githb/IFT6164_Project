# **Training Set**

```python
import projector
import random
from os.path import isfile, join

for c in ["ship", "truck"]: #["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]:
    flaglst = [False for i in range(5000)]
    subdirs = os.listdir("out_train/"+c)
    missing = len(subdirs)
    for num in subdirs:
        flaglst[int(num)-1] = True
    for f in range(0,250-missing):
        x = random.randint(1, 5000)
        while flaglst[x-1] == True:
            x = random.randint(1, 5000)
        flaglst[x-1] = True
        fname = str(x).zfill(4)
        projector.run_projection("cifar10.pkl", "cifar10/train/"+c+"/"+fname+".png", "out_train/"+c+"/"+fname, False, 303, 1000)
```
